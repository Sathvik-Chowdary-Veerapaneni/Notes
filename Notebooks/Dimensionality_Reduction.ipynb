```python
import os
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Image Path

```python
image_path='/Users/sathvikchowdaryveerapaneni/Desktop/GitHub/Machine_Learning_101/Image_Classification_101/Animals/Lion/africa-african-animal-cat-41315.jpeg'
```

```python
# Reading the image and storing in a variable
RGB_Image=mpimg.imread(image_path)

#Displaying the color image
plt.imshow(RGB_Image)
```

```python
#Shape of the image
RGB_Image.shape
```

```python
# Function to calculate the number of featuers in a image, H*W*C (Height * Width * Channels) 
# It will provide number of pixels and each pixel is an feature


def cal_num_featuers(img):
    temp=1
    for i in img.shape:
        temp*=i
    return temp
```

```python
# Number of featuers in the RGB_Image
num_featuers=cal_num_featuers(RGB_Image)
print(num_featuers)
```

```python
# Plotting the image labeling its title and shape
# Labeling num of featuers in the x-axis

plt.title('RGB Image'+ str(RGB_Image.shape))
plt.xlabel('Number of featuers' + str(num_featuers))
plt.imshow(RGB_Image)
```

```python
# Converting RGB image to gray scale. 
# Observe what we getting over here.
gray_image = cv2.cvtColor(RGB_Image,cv2.COLOR_RGB2GRAY)
gray_image.shape
num_featuers_gray=cal_num_featuers(gray_image)
```

```python
plt.title('Gray Image'+str(gray_image.shape))
plt.xlabel('Number of featuers '+ str(num_featuers_gray))
plt.imshow(gray_image)
```
